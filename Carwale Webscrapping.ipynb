```python
#importing required libraries

import pandas as pd
import selenium
from selenium import webdriver
import time
import requests
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
```

```python
#loading the drivers and url
driver=webdriver.Chrome(r'Chromedriver.exe')
driver.maximize_window() #To maximise the window

#getting targeted url
driver.get('https://www.carwale.com/used/cars-in-delhi/#city=10&pc=10&sc=-1&so=-1&pn=1')
```

```python
#creating an empty lists

names=[]
model=[]
Kilometers=[]
Fuel=[]
Location=[]
EMI=[]
Price=[]


for i in range(0,50):
    driver.execute_script("window.scrollBy(0,50000)")
    names_Tag=driver.find_elements(By.XPATH,'//span[@class="spancarname card-detail-block__title-text-with-year"]')
    for i in names_Tag:
        try:
            names.append(i.text)# saving in the list
        except:
            names.append(i.text)("NA")
    
    model_Tag=driver.find_elements(By.XPATH,'//span[@class="slYear card-detail-block__title-text-with-year"]')
    for i in model_Tag:
        try:
            model.append(i.text)# saving in the list
        except:
            model.append(i.text)("NA")

    KM_Tag=driver.find_elements(By.XPATH,"//span[@class='slkms']")
    for i in KM_Tag:
        try:
            Kilometers.append(i.text)# saving in the list
        except:
            Kilometers.append(i.text)("NA")

    Fuel_Tag=driver.find_elements(By.XPATH,"//span[@class='fuel']")
    for i in Fuel_Tag:
        try:
            Fuel.append(i.text)# saving in the list
        except:
            Fuel.append(i.text)("NA")

    Location_Tag=driver.find_elements(By.XPATH,"//span[@class='cityName']")
    for i in Location_Tag:
        try:
            Location.append(i.text)# saving in the list
        except:
            Location.append(i.text)("NA")

    EMI_Tag=driver.find_elements(By.XPATH,"//p[@class='card-detail__emi-text']")
    for i in EMI_Tag:
        try:
            EMI.append(i.text)# saving in the list
        except:
            Kilometers.append(i.text)("NA")
        
    Price_Tag=driver.find_elements(By.XPATH,'//span[@data-bind="text: Price"]') 
    for i in Price_Tag:
        try:
            Price.append(i.text)# saving in the list
        except:
            Price.append(i.text)("NA")    
    print(len(names),len(model),len(Kilometers),len(Fuel),len(Location),len(EMI),len(Price))
```

```python
df = pd.DataFrame({"Car Name":names[:10000],"Year":model[:10000],"Kilometers":Kilometers[:10000],"Fuel":Fuel[:10000],"Location":Location[:10000],"EMI":EMI[:10000],"Price":Price[:10000]})
df                
```

```python
df.to_csv('Carwale Used Cars data.csv')
```
